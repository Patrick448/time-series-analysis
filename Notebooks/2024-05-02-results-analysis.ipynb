```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
def plot_error_by_timestep(path, column_name, label=None):
    df = pd.read_csv(path)
    string_list = df[column_name].tolist()
    list_split = [x.split(',') for x in string_list]
    val_list = [[float(x) for x in y] for y in list_split]
    val_list = np.array(val_list)
    plt.plot(val_list.mean(axis=0), label=label)
    plt.legend()

   


def all_errors_by_timestep(path):
    plot_error_by_timestep(path, 'RMSE_by_timestep', 'RMSE')
    plot_error_by_timestep(path, 'MAE_by_timestep', 'MAE')
    plot_error_by_timestep(path, 'MAPE_by_timestep', 'MAPE')
    plot_error_by_timestep(path, 'MSE_by_timestep', 'MSE')
    plt.show()

all_errors_by_timestep('../experiments/exp5/result.csv')
```
