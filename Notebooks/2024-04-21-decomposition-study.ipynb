```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv('../processed_data/price_weather_weekly_df-2016-2023.csv')
df.set_index('dt', inplace=True)
df.index = pd.to_datetime(df.index)
df['Alface Crespa - Roça'].plot(figsize=(15,6))
plt.show()
```

```python
res = sm.tsa.seasonal_decompose(df['Alface Crespa - Roça'])
res.plot()
plt.show()
```

```python
res.resid.plot(figsize=(15,6))
df['Alface Crespa - Roça'].plot(figsize=(15,6))
plt.show()
```
