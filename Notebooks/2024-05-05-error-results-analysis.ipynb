```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import csv
```

```python
def plot_error_by_experiment_group(path, column_name, title):
    df = pd.read_csv(path)
    ids = df['experiment_group'].unique().tolist()
    for i in ids:
        df_i = df[df['experiment_group'] == i]
        string_list = df_i[column_name].tolist()
        list_split = [x.split(',') for x in string_list]
        val_list = [[float(x) for x in y] for y in list_split]
        val_list = np.array(val_list)
        plt.plot(val_list.mean(axis=0), label=i, marker='o')
        plt.title(title)
        plt.legend()
    plt.show()

plot_error_by_experiment_group('../experiments/exp6_2024-05-05/result.csv', 'RMSE_by_timestep', 'RMSE')
plot_error_by_experiment_group('../experiments/exp6_2024-05-05/result.csv', 'MAE_by_timestep', 'MAE')
plot_error_by_experiment_group('../experiments/exp6_2024-05-05/result.csv', 'MAPE_by_timestep', 'MAPE')
plot_error_by_experiment_group('../experiments/exp6_2024-05-05/result.csv', 'MSE_by_timestep', 'MSE')
plot_error_by_experiment_group('../experiments/exp6_2024-05-05/result.csv', 'R2_by_timestep', 'R2')
```
