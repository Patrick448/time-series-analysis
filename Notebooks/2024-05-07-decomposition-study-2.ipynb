```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import random
```

```python
df = pd.read_csv('../processed_data/price_weather_weekly_df-2016-2023.csv')
df.set_index('dt', inplace=True)
df.index = pd.to_datetime(df.index)

values = []
for i in range(len(df)):
    values.append(random.uniform(0,10))

rand_df = pd.DataFrame(values)
rand_df.index = df.index
rand_df.plot()
plt.show()
res = sm.tsa.seasonal_decompose(rand_df)
res.plot()
plt.show()
res.resid.plot(figsize=(15, 6))
```

```python
values = []
for i in range(len(df)):
    next_val = 0 if len(values) == 0 else values[-1] + random.uniform(0,1)
    values.append(next_val)

rand_df = pd.DataFrame(values)
rand_df.index = df.index
rand_df.plot()
plt.show()
res = sm.tsa.seasonal_decompose(rand_df)
res.plot()
plt.show()
res.resid.plot(figsize=(15, 6))
```

```python
values = []
for i in range(len(df)):
    next_val = 0 if len(values) == 0 else values[-1] + random.uniform(-1,1)
    values.append(next_val)

rand_df = pd.DataFrame(values)
rand_df.index = df.index
rand_df.plot()
plt.show()
res = sm.tsa.seasonal_decompose(rand_df)
res.plot()
plt.show()
res.resid.plot(figsize=(15, 6))
```
